# Doğrusal İlişkinin Gösterilmesi

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
df = tips.copy()
df.head()
```

```python
sns.lmplot(data=df, x='total_bill', y='tip');
# Doğru etrafındaki açık renkli alan sapmayı ifade ediyor.
# Hesap ödemesi arttıkça bahşiş miktarı da artıyor.
# Fakat x düzleminde 30-50 değerleri arasında standart sapma oranı oldukça artmış.
```

```python
sns.lmplot(data=df, x='total_bill', y='tip', hue='smoker');
# Grafikte görüleceği üzere sigara kullanmayanların ödedikleri bahşiş miktarı; sigara kullananlara göre daha fazla.
# Yani sigara içmeyenler daha fazla bahşiş veriyorlar.
```

```python
sns.lmplot(data=df, x='total_bill', y='tip', hue='smoker', col='time');
# sabah da akşam da sigara içmeyenlerin hesap ücreti arttıkça bahşiş verme oranları artıyor.
# Akşam daha fazla trafik olduğu gözüküyor.
# Akşam sigara kullananların standart sapması oldukça yüksek.
```

```python
sns.lmplot(data=df, x='total_bill', y='tip', hue='smoker', col='time', row='sex');
```
