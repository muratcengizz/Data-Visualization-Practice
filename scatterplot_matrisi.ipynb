# Scatterplot Matrisi
## Scatterplot sayısal değişkenler arasındaki ilişkiyi belirten saçılım grafiğidir!

```python
import seaborn as sns
```

```python
# Aşağıdaki veri setinde bazı çiçek türleri var. Bu türlere ait sayısal özellikleri belirten değişkenler var
```

```python
iris = sns.load_dataset('iris')
df = iris.copy()
df.head()
```

```python
df.describe().T
# 150 adet gözlem var
# sayısal değişkenlerin ortalama ve standart sapma değerleri minimum ve maksimum değerlerde göz önüne alınarak gayet iyi.
```

```python
df.info()
# 4 adet sayısal sürekli değişken var. 1 adet kategorik değişken var.
```

## Kategorik değişken içerisinde sıralama olmadığı için veri tipini object -> categorical 'a çevirmemize gerek yok!

```python
df.dtypes
# 4 adet sayısal değişken, 1 adet kategorik değişken var
```

```python
sns.pairplot(df);
# Eğer 2 değişkenli grafiklerde toz bulutu yapısı yok ise, bu iki değişkenin birbiri ile karşılaştırılmasının doğru olmadığı anlamına gelir.
# doğru olanlara örnek: y düzleminde ki petal_length ile x düzleminde ki petal_width değişkenlerinin kesişimleri
# yanlış olanlara örnek: y düzleminde ki sepal_length ile x düzlemindeki petal_length değişkenlerinin kesişimleri
```

```python
sns.pairplot(df, hue='species');
```

```python
sns.pairplot(data=df, kind='reg');
```

```python
sns.pairplot(data=df, kind='reg', hue='species');
```
