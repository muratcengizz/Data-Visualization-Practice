# Korelasyon Grafikleri

## Scatterplot
### total_bill: yemeğin toplam fiyatı (bahşiş ve vergi dahil)
### tip: bahşiş
### sex: ücreti ödeyen kişinin cinsiyeti (0=male, 1=female)
### smoker: grupta sigara içen var mı ? (0=no, 1=yes)
### day: gün
### time: Ne zaman ? (0=day, 1=night)
### size: grupta kaç kişi var ?

```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
df = tips.copy()
df.head()
```

```python
sns.scatterplot(data=df, x='total_bill', y='tip');
```

## Korelasyon Çaprazlamalar

```python
sns.scatterplot(data=df, x='total_bill', y='tip', hue='time');
```

```python
sns.scatterplot(data=df, x='total_bill', y='tip', hue='time', style='time');
```

```python
sns.scatterplot(data=df, x='total_bill', y='tip', hue='day');
```

```python
sns.scatterplot(data=df, x='total_bill', y='tip', hue='day', style='time');
```

```python
sns.scatterplot(data=df, x='total_bill', y='tip', hue='sex');
```

```python
sns.scatterplot(data=df, x='total_bill', y='tip', hue='time', style='sex');
```

```python
sns.scatterplot(data=df, x='total_bill', y='tip', size='size');
```

```python
sns.scatterplot(data=df, x='total_bill', y='tip', hue='size', size='size');
```
