# Displot Practice

```python
import seaborn as sns
from pandas.api.types import CategoricalDtype
```

```python
diamonds = sns.load_dataset('diamonds')
df = diamonds.copy()
df.head()
```

```python
cut_siralama = ['Fair', 'Good', 'Very Good', 'Premium', 'Ideal']
df.cut = df.cut.astype(CategoricalDtype(categories=cut_siralama, ordered=True))
df.cut
```

```python
sns.displot(df.price);
# histogram grafiği çizdik.
```

```python
sns.displot(data=df, x='cut');
```

```python
sns.displot(data=df, x='table', kind='kde');
# yoğunluk grafiği
```

```python
sns.displot(data=df, x='cut', kind='ecdf');
# yoğunluk grafiği
```

```python
sns.displot(data=df, x='cut', y='price');
```

```python
sns.displot(data=df, x='table', y='price', kind='kde', kde=True);
```

```python
sns.kdeplot(df.price, shade=True);
# shade parametresi 'fill' yani doldur anlamına gelir. shade=True için, yoğunluk grafiğinin içerisi doldurulur.
```

```python
sns.kdeplot(df.price);
```

```python
sns.displot(df.price);
```

```python
sns.displot(data=df, x='price', hue='color');
```

```python
sns.displot(data=df, x='price', hue='color', multiple='stack');
```

```python
sns.displot(data=df, x='price', hue='color', height=8, aspect=4);
```
