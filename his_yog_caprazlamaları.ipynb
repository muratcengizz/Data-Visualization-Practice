```python
import seaborn as sns
```

```python
diamonds = sns.load_dataset('diamonds')
df = diamonds.copy()
df.head()
```

```python
sns.kdeplot(df.price, shade=True);
# shade=True parametresi; yoÄŸunluk grafiÄŸinin iÃ§ini doldurur. Default olarak False 'dur.
```

```python
sns.displot(data=df, x='price', kind='kde');
```

```python
(sns
 .FacetGrid(df,
           hue='cut',
           height=5,
           xlim=(0,10000))
 .map(sns.kdeplot, 'price', shade=True)
 .add_legend()
);
# FacetGrid fonksiyonu; grafik Ã¼zerine dÃ¼ÅŸen boyutlarÄ± keserek gÃ¶stermek iÃ§in kullanÄ±lÄ±r.
# hue = boyut ekleme parametresi
# xlim = x ekseninde yer alacak deÄŸiÅŸkenin hangi aralÄ±klarda yer alacaÄŸÄ±nÄ± belirtebiliriz.
```
