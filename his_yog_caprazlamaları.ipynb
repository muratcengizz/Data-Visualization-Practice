# Çaprazlamalar

```python
import seaborn as sns
```

```python
diamonds = sns.load_dataset('diamonds')
df = diamonds.copy()
df.head()
```

```python
sns.kdeplot(df.price, shade=True);
# shade=True parametresi; yoğunluk grafiğinin içini doldurur. Default olarak False 'dur.
```

```python
sns.displot(data=df, x='price', kind='kde');
```

```python
(sns
 .FacetGrid(df,
           hue='cut',
           height=5,
           xlim=(0,10000))
 .map(sns.kdeplot, 'price', shade=True)
 .add_legend()
);
# FacetGrid fonksiyonu; grafik üzerine düşen boyutları keserek göstermek için kullanılır.
# hue = boyut ekleme parametresi
# xlim = x ekseninde yer alacak değişkenin hangi aralıklarda yer alacağını belirtebiliriz.
```

```python
#sns.FacetGrid(df, hue='cut', height=5, xlim=(0,10000));
```

```python
sns.FacetGrid(df, hue='cut', height=10, xlim=(0,10000)).map(sns.kdeplot, 'price', shade=True).add_legend();
# height parametresi grafiğin yüksekliğini belirtir.
# xlim parametresi x paralelinde yer alacak değişkenin hangi aralıklarda yer alması gerektiğini belirtir.
```

```python
sns.catplot(data=df, x='cut', y='price', hue='color', kind='point');
```
