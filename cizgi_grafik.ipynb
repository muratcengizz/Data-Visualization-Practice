# Çizgi Grafik

```python
import seaborn as sns
```

```python
fmri = sns.load_dataset('fmri')
df = fmri.copy()
df.head()
```

### subject: verilerin toplandığı kişiler
### timepoint: zaman noktaları
### event: verinin toplanmasıyla ilgili birbirinden farklı olayları ifade ediyor.
### region: sinyalin toplandığı bölgeyi ifade ediyor.
### signal: sinyali ifade ediyor.

```python
df.describe().T
```

```python
df.info()
# 1064 adet gözlem var.
# 5 adet değişken var.
# 3 adet kategorik değişken, 2 adet sayısal değişken var.
```

```python
df.columns
# değişkenleri listeledik.
```

```python
df.dtypes
# değişkenlerin yapısını listeledik.
```

```python
df.head()
```

```python
df[['timepoint']].describe().T
# 1064 adet gözlem var.
# ortalaması 9
# standart sapması 5.4
# minimum değeri 0, maksimum değeri 18.
```

```python
df[['signal']].describe().T
# standart sapma, ortalamaya göre yüksek.
# signal değişkenine ait verilerin çoğunluğu maksimum değerden uzak.
```

```python
df.groupby('timepoint')[['signal']].describe()
# timepoint değişkenine göre signal değişkeninin temel bilgilerini edindik.
# timepoint değişkenine ait 0-18 arasında sınıf var. Her sınıfta 56 gözlem var.
# minimum ve maksimum değerleri bakacak olursak ince çalışılması gerekilen bir veri seti. 
# Ve büyük ihtimalle insanlar tarafında oluşturulmamış, aksine makineler tarafından oluşturulmuş gibi.
# standart sapma değerleri ortalamadan fazla.
```

```python
df.groupby('subject')['subject'].count()
# subject değişkenine ait s0 'dan s13 'e kadar toplam 13 tane sınıf var. Ve her sınıfa ait 76 gözlem var.
```

```python
df.groupby('timepoint')[['signal']].count().T
# timepoint değişkenine ait 0'dan 18'e toplam 18 adet sınıf var. Her sınıfta 56 adet signal gözlemi var.
# her zaman noktasında eşit sayıda(56) sinyal olması bu verilerin mekanik veri olduğunu gösteriyor.
```

```python
df.groupby('signal').count()
# her sinyal eşsiz. 
```

```python
df.groupby('timepoint')['signal'].describe()
```

```python
df.groupby(['subject', 'region'])['signal'].describe()
```

# Çizgi Grafik ve Çaprazlamalar

```python
sns.lineplot(data=df, x='timepoint', y='signal');
```

```python
sns.lineplot(data=df, x='timepoint', y='signal', hue='event');
# yukarıdaki gibi pig(sıçrama) olmasının sebebi event değişkenine ait stim sınıfı imiş.
```

```python
sns.lineplot(data=df, x='timepoint', y='signal', hue='event', style='event');
```

```python
sns.lineplot(data=df,
            x='timepoint',
            y='signal',
            hue='event',
            style='event',
            markers=True);
```

```python
sns.lineplot(data=df,
            x='timepoint',
            y='signal',
            hue='event',
            style='event',
            markers=True,
            dashes=False);
```

```python
sns.lineplot(data=df,
            x='timepoint',
            y='signal',
            hue='region',
            style='event');
```
